* The chart should allow you to easily visuallize both the **total spending** by customers in a group on all product categories and the **composition** of their spending on those product categories.

ii. Further, answer the following questions in a **text (markdown) cell** and make sure to explain your answer based on the chart:

**Hint**: By default, each row in the dataset is shown as a separate rectangle in the bar, and these rectangles have borders that can make the chart look cluttered. In that case, use `fig.update_traces(opacity=1, marker_line_width=0)` to remove the border lines.

```python
# Task B.1 â€“ Spending by Education Level (stacked bar chart)

import plotly.express as px
import pandas as pd

# List of product spending columns
spend_cols = [
    "MntWines","MntFruits","MntMeatProducts",
    "MntFishProducts","MntSweetProducts","MntGoldProds"
]

# Melt the data to long format: each row = Education + Product + Spend
long_ed = df.melt(
    id_vars=["Education"],
    value_vars=spend_cols,
    var_name="Product",
    value_name="Spend"
)

# Group totals by Education & Product
ed_grp = long_ed.groupby(["Education","Product"], as_index=False)["Spend"].sum()

# Create stacked bar: total spending by education, colored by product
fig = px.bar(
    ed_grp,
    x="Education",
    y="Spend",
    color="Product",
    title="Total Spending by Education Level (Stacked by Product Category)"
)

# Remove inner borders to declutter the chart
fig.update_traces(opacity=1, marker_line_width=0)
fig.show()
```

Customers with a Graduation education level contribute the highest total revenue across all product categories.

Within this group, wines and meat products make up the largest portions of total spending, indicating a preference for higher-value product categories.

Customers with Masterâ€™s or PhD education levels also show moderate spending, while Basic and 2n Cycle education groups contribute comparatively less.

Overall, education level appears positively related to total spending, suggesting that higher-educated customers are more valuable to the company.

```python
# Task B.2 â€“ Spending by Marital Status (grouped bar chart)

# Melt again for Marital_Status
long_ms = df.melt(
    id_vars=["Marital_Status"],
    value_vars=spend_cols,
    var_name="Product",
    value_name="Spend"
)

# Group totals by Marital Status & Product
ms_grp = long_ms.groupby(["Marital_Status","Product"], as_index=False)["Spend"].sum()

# Create grouped bar chart: each marital group shows spending per product category
fig = px.bar(
    ms_grp,
    x="Marital_Status",
    y="Spend",
    color="Product",
    barmode="group",
    title="Spending by Marital Status (Grouped by Product Category)"
)

# Improve readability
fig.update_traces(opacity=1, marker_line_width=0)
fig.show()
```

Across all marital-status groups, wine products account for the highest spending, followed by meat and gold products, indicating these are the most valued categories overall.

The married and together (in-relationship) groups show the largest total expenditures, with noticeably higher bars for wines and meats compared to other groups.

Single and divorced customers spend less overall but still follow a similar product-preference pattern.

Marital status clearly influences spending distribution â€” customers who share households (married or together) tend to purchase in larger quantities and spend more on premium categories, while those living alone allocate smaller budgets across product types.

## Part C (30 points)

In this part, let us investigate variability in customers' online purchasing behavior.

i. We will use a visualization chart that shows distribution of customers' online purchases. The chart should:
* Show multiple distributions, each being the distribution of number of online purchases by customers of a certain generation (Gen X vs. Millennials,  ignore other generations)
* Allow users to observe the 1st, 2nd, and 3rd quartiles of the distribution, along with any outliers.

ii. Based on the chart, answer the following question:
* How would you comment on the pattern of online purchases across different generations.

iii. In your chart, further divide each distribution into subgroups of customers based on the **number of kids at home** (not including teens).
How would you comment on the pattern of online purchases across the different subgroups of customers?




```python
# Part C â€“ Online Purchase Distributions (Box Plots)

import plotly.express as px

# Helper function to classify generations based on Year_Birth
def get_generation(year):
    if 1965 <= year <= 1980:
        return "Gen X"
    elif 1981 <= year <= 1996:
        return "Millennials"
    else:
        return "Other"

# Creating a Generation column
df["Generation"] = df["Year_Birth"].apply(get_generation)

# Keeping only Gen X and Millennials for analysis
df_gen = df[df["Generation"].isin(["Gen X", "Millennials"])].copy()

# Step 1: Compare online purchases across generations
fig = px.box(
    df_gen,
    x="Generation",
    y="NumWebPurchases",
    points="outliers",
    color="Generation",
    title="Online Purchases by Generation (Box Plot showing Quartiles and Outliers)"
)

# Decouple shared y-axis scaling for clarity (per Assignment 1 feedback)
fig.update_yaxes(matches=None)
fig.show()

# Step 2: Add sub-groups by number of kids at home (Kidhome only, not teens)
fig = px.box(
    df_gen,
    x="Generation",
    y="NumWebPurchases",
    color="Generation",
    facet_col="Kidhome",          # creates separate panels for each Kidhome count
    facet_col_wrap=3,
    points="outliers",
    title="Online Purchases by Generation, Split by Number of Kids at Home"
)

# Keep each facet on its own scale to avoid distortion
fig.update_yaxes(matches=None)
fig.for_each_annotation(lambda a: a.update(text=a.text.replace("Kidhome=", "Kids at home = ")))
fig.show()
```

Pattern for online purchases across different generations:

Millennials generally make slightly more online purchases than Gen X, with higher median values and more frequent outliers visible in the box plot.
This suggests Millennials are somewhat more engaged in online shopping, possibly reflecting their stronger familiarity with digital platforms and e-commerce habits.

Gen X customers display a wider spread of purchase frequencies, with some heavy online buyers but also many low-activity users.

Effect of number of kids at home:

When divided by Kidhome groups, customers with no kids (Kidhome = 0) show the highest online purchase activity for both generations.
Those with 1 or 2 kids make fewer purchases on average, and their distributions are tighter, indicating less variability.
This pattern suggests that family responsibilities and household size may limit available time or discretionary income for online shopping, especially among Gen X households.
Millennials without children remain the most active online buyers overall.
